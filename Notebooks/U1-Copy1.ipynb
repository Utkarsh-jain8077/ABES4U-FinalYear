```python
def Personality(a):
    classes=["ESTJ","ENTJ","ESFJ","ENFJ","ISTJ","ISFJ","INTJ","INFJ","ESTP","ESFP","ENTP","ENFP","ISTP","ISFP","INTP","INFP"]
    s=""
    if(a[0]+a[1]+a[2]>=2):
        s+='E'
    else:
        s+='I'
    if(a[3]+a[4]+a[5]>=2):
        s+='S'
    else:
        s+='N'
    if(a[6]+a[7]+a[8]>=2):
        s+='T'
    else:
        s+='F'
    if(a[9]+a[10]+a[11]>=2):
        s+='J'
    else:
        s+='P'
    return classes.index(s)+1
```
