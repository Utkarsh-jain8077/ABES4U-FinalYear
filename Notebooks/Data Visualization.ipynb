```python
def tenth(a):
    if a<=10:
        return a*9.5
    else:
        return a

def elev(a):
    if a<10:
        return 75.0
    else:
        return a
```

```python
ad1["10th"]=ad1["10th"].apply(tenth)
# ad["11th"]=ad["11th"].apply(elev)
# ad1["12th"]=ad1["12th"].apply(elev)
```

```python
ad1["11th"]=ad1["11th"].apply(elev)
```

```python
ad1["12th"]=ad1["12th"].apply(elev)
```

```python
ad1["12th"].value_counts()
```

```python
ad1.head()
```

```python
def Personality(a):
    classes=["ESTJ","ENTJ","ESFJ","ENFJ","ISTJ","ISFJ","INTJ","INFJ","ESTP","ESFP","ENTP","ENFP","ISTP","ISFP","INTP","INFP"]
    s=""
    if(a[0]+a[1]+a[2]>=2):
        s+='E'
    else:
        s+='I'
    if(a[3]+a[4]+a[5]>=2):
        s+='S'
    else:
        s+='N'
    if(a[6]+a[7]+a[8]>=2):
        s+='T'
    else:
        s+='F'
    if(a[9]+a[10]+a[11]>=2):
        s+='J'
    else:
        s+='P'
    return classes.index(s)+1
```

```python
v1=ad1.iloc[:,4:-1]
```

```python
v1
```

```python
ad1["personality"]=v1.apply(Personality,axis=1)
```

```python
ad1.head()
```

```python
ad1.drop(["EvsIn(1)","EvsIn(2)","EvsIn(3)","SvsI(1)","SvsI(2)","SvsI(3)","TvsF(1)","TvsF(2)","TvsF(3)","JvsP(1)","JvsP(2)","JvsP(3)"],axis=1,inplace=True)
```
